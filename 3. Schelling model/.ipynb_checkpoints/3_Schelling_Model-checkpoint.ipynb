```python
import os 
import numpy as np
from tqdm.notebook import tqdm

percents = np.linspace(0, 1, 9)

num_of_players = 2
n = 100
num_of_iterations = 200

# schelling = Schelling(num_of_players=num_of_players, n=n, r=0.125)

# for i in range(num_of_iterations):
#     schelling.play()

schellings = []
for percent in tqdm(percents):
    schelling = Schelling(num_of_players=num_of_players, n=n, r=percent)

    for i in range(num_of_iterations):
        schelling.play()

    schellings.append(schelling)
  # schelling.save_image(num_of_iteration=i, save_img=True)

# path = '/content/drive/My Drive/Schelling model'
# [os.mkdir(os.path.join(path, f'{r}_percent_of_agents_folder')) for r in percents]


```

```python
plot_want_to_move(schellings, num_of_iterations)
```
