```python
from julia import julia_set, plot_julia
```

```python
julia, c = julia_set(c=-0.5 + 0.9j, num_iterations=300)
plot_julia(julia, c, title=True, save_img=False, color=True)
```

```python
r = (1 + 5**0.5) / 2 
c = 1 - r

julia, c = julia_set(c=c)
plot_julia(julia, c, title=True, save_img=False, color=True)
```
